## Finding Correlation in data we are using

```python
df_copy = df.copy()


distance_labels = ['0-414', '415-586', '587-719', '720-844', '845-1032', '1033-1289', '1290-1744', '1745-4983']

df_copy['Distance Group'] = pd.cut(df_copy['Flight Distance'],
                                    bins=[0, 414, 586, 719, 844, 1032, 1289, 1744, 4983],
                                    labels=distance_labels
                                    )
plt.figure(figsize=(20,14))
sns.histplot(data=df_copy,x='Distance Group',hue='Distance Group',zorder=5,alpha=1,palette='Set1')
plt.title('Satisfaction  by Flight Distance', fontsize=22)
plt.xlabel('Flight Distance', fontsize=16)
plt.ylabel('Satisfaction', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```
